```python
li = [2 ,0, "hello", None]


for element in li:
  try:
    print("current element: ",element)
    print(5/element) 
  except:
    print("Error Occured")
  print("-"*35)
```

```python
li = [2 ,0, "hello", None]


for element in li:
  try:
    print("current element: ",element)
    print(5/element) 
  except Exception as e:
    print("Error Occured: ", str(e))
    
  print("-"*35)
```

```python
li = [2 ,0, "hello", None]


for element in li:
  try:
    print("current element: ",element)
    print(5/int(element))
  except ZeroDivisionError:
    print("Element is Zero")
  except Exception as e:
    print("Error Occured: ", str(e))
    
  print("-"*35)
```

```python
li = [2 ,0, "hello", None]


for element in li:
  try:
    print("current element: ",element)
    print(5/int(element))

  except Exception as e:
    print("Error Occured: ", str(e))

  except ZeroDivisionError:
    print("Element is Zero")

  except NoneTypeError: 
    print("None elemetn")
  
    
  print("-"*35)
```
